## Background

The SAT and ACT are standardized tests that many colleges and universities in the United States require for their admissions process. This score is used along with other materials such as grade point average (GPA) and essay responses to determine whether or not a potential student will be accepted to the university.

The SAT has two sections of the test: Evidence-Based Reading and Writing and Math ([*source*](https://www.princetonreview.com/college/sat-sections)). The ACT has 4 sections: English, Mathematics, Reading, and Science, with an additional optional writing section ([*source*](https://www.act.org/content/act/en/products-and-services/the-act/scores/understanding-your-scores.html)). They have different score ranges, which you can read more about on their websites or additional outside sources (a quick Google search will help you understand the scores for each test):
* [SAT](https://collegereadiness.collegeboard.org/sat)
* [ACT](https://www.act.org/content/act/en.html)

Standardized tests have long been a controversial topic for students, administrators, and legislators. Since the 1940's, an increasing number of colleges have been using scores from sudents' performances on tests like the SAT and the ACT as a measure for college readiness and aptitude ([*source*](https://www.minotdailynews.com/news/local-news/2017/04/a-brief-history-of-the-sat-and-act/)). 

In addition, SAT and ACT have been close substitutes for each other for many years. Up until 2012, the SAT's popularity exceeeded that of the ACT - it was 55% versus 45% in market share terms. However, as the perception that colleges had a preference for SAT over ACT waned, ACT began eroding SAT's market share. In 2012, the ACT overtook SAT in market share. This was in part due to many states adopting the ACT as their official state test for juniors ([*source*](https://theolivebook.com/sat-vs-act-which-is-more-popular/#:~:text=As%20of%202019%2C%20the%20SAT,students%20who%20took%20the%20ACT.)).

The SAT took notice of the decline and amended their format in March 2016. Since then, SAT's participation rate has been on the rise and surpassed ACT in 2019. The important question before the College Board (the organisation that administers the SAT) is then to devise strategies to maintain its competitive edge over ACT.

```python
# Setting out a function to calculate mean:
def mal_mean(list_of_num):
    return sum(list_of_num) / len(list_of_num)

# Setting out a list of numbers to check that the defined function works:
list_of_num = [1, 2, 3, 4, 5]
print(mal_mean(list_of_num))
```

2. Manually calculate standard deviation:

    The formula for standard deviation is below:

    $$\sigma = \sqrt{\frac{1}{n}\sum_{i=1}^n(x_i - \mu)^2}$$

    Where $x_i$ represents each value in the dataset, $\mu$ represents the mean of all values in the dataset and $n$ represents the number of values in the dataset.

    Write a function that takes in values and returns the standard deviation of the values using the formula above. Hint: use the function you wrote above to calculate the mean! Use the list of numbers you created above to test on your function.
    
    *Note*: Do not use any standard deviation methods built-in to any Python libraries to do this! This should be done without importing any additional libraries.

```python
# Setting out a function to calculate standard deviation:
def standard_deviation(list_of_num):
    summation = 0
    for i in list_of_num:
        summation += ((i-mal_mean(list_of_num))**2)
    return (((1/len(list_of_num))*(summation))**(1/2)) 
    
# Using the list of numbers we defined earlier to check that the defined function works:
print(standard_deviation(list_of_num))
```

```python
#  Creating a custom function using plt.subplot to draw histograms"
def subplot_histograms(dataframe, list_of_columns, list_of_titles, list_of_xlabels,
                      list_of_ylabels, figsize, xlim, ylim):
    nrows = int(np.ceil(len(list_of_columns)/3))  # using the numpy ceiling function to make sure that we can enough rows
    fig, ax = plt.subplots(nrows=nrows, ncols=3, figsize = figsize) #ncols = 3 since we have 3 years each for SAT and ACT
    ax = ax.ravel() # Ravel works just like reshape(-1)
    
    for i, column in enumerate(list_of_columns):
        ax[i].hist(dataframe[column]) 
        ax[i].set_title(list_of_titles[i], fontsize = 15)    
        ax[i].set_xlabel(list_of_xlabels[i], fontsize = 12)
        ax[i].set_ylabel(list_of_ylabels[i], fontsize = 12)
        ax[i].set_xlim(xlim[i])
        ax[i].set_ylim(ylim[i])

# References:
# https://github.com/Phlya/adjustText/issues/58
# https://adjusttext.readthedocs.io/en/latest/Examples-for-multiple-subplots.html
# https://stackoverflow.com/questions/50455551/why-is-numpy-ravel-required-in-this-code-that-produces-small-multiples
```
